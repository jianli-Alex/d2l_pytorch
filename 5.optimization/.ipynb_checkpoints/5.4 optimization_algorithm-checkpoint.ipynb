## 优化与深度学习

在DL中，通常定义一个损失函数，然后转化为一个优化问题，但是优化和深度学习是有一些不同的，前者是基于训练集，其通常是为了降低`训练误差`，而深度学习的是目标是降低`泛化误差`

在许多优化问题上，我们希望找到全局的最优点。但是对于深度学习问题，由于损失函数非凸性，存在着大量的驻点，我们很难寻找到一个全局的最优点
- 然后，在深度学习中，陷入局部最优点所带来的问题往往无关紧要，通常来说局部最优点和全局最优点的结果可能相差无几。实际上，我们通常想要寻找的是平坦的局部最优点（有更强的鲁棒性）
- 但是对于深度学习中，大量的驻点基本都是鞍点，所以我们需要做的是逃离鞍点
    - 若一个函数的输入向量维度为k，输出是一个标量，那么它的海塞矩阵有k个特征值，其中当所有特征值为`正`时，该驻点为`局部最小值`，当所有特征值都是`负`时，该驻点为`局部最大值`，当特征值中`既有正又有负`时，该驻点为`鞍点`。
    - 根据随机矩阵理论，对于一个大的高斯随机矩阵来说，出现正和负的特征值都为0.5，显然基本上所有的驻点都是鞍点

### 局部最小值

以函数$f(x) = xcos(\pi x)$为例

```python
x = np.linspace(-1, 2, 1000)
y = x*np.cos(np.pi*x)
plt.plot(x, y)
# 添加注释
plt.annotate("局部最小值", xy=(-0.3, -0.25), xytext=(-0.3, -0.75), arrowprops={"arrowstyle":"->"})
plt.annotate("全局最小值", xy=(1.1, -0.95), xytext=(0.9, 0.5), arrowprops={"arrowstyle":"->"})
plt.xlabel("x")
plt.ylabel("f(x)")
```

## 优化算法

### 一维梯度下降

通常来说，在深度学习中，比较少用批量梯度下降。因为数据量往往很大，无法一次遍历，或者需要花费很长时间才能进行更新。通常采用小批量梯度下降，而随机梯度下降由于采用单个样本进行更新，实际上无法充分利用GPU资源进行并行运算

给定一个绝对值很小的值$\epsilon$，根据泰勒展开式子有：$f(x+\epsilon) = f(x) + \epsilon f'(x)$

然后，找一个很小的常数$\eta$（学习率）,使得$|\eta f'(x)|$足够小，并用$|\eta f'(x)|$代替$\epsilon$，最后得到$f(x - \eta f'(x)) = f(x) - \eta f'(x)^2$，当$f'(x) \neq 0$的时候，有$f(x - \eta f'(x)) \leq f(x)$

即使用$x := x - \eta f'(x)$，函数f(x)的值可能会降低

```python
# f(x) = x^2为例
def gd(eta):
    # 初始值
    x = 10
    # 存储每一次的更新值
    result = [x]
    # 更新10次
    for i in range(10):
        x -= eta * 2 * x
        result.append(x)
        
    print(f"epoch 10, x: {x}")
    return result

result = gd(0.2)
```

```python
# 画出x的迭代轨迹
def show_trace(result):
    # 取最大值用于画图
    n = max(abs(min(result)), abs(max(result)), 10)
    # 坐标点
    x_line = np.arange(-n, n, 0.1)
    # 画图
    plt.plot(x_line, x_line**2)
    plt.plot(result, list(map(lambda x: x**2, result)), "-o")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    
show_trace(result)
```

#### 学习率

不同的学习率会导致不同的效果
- 太小时更新缓慢
- 太大时无法收敛

```python
# 学习率为0.03
result = gd(0.03)
show_trace(result)
```

```python
# 学习率为1.5
result = gd(1.1)
show_trace(result)
```

### 多维梯度下降

如果输入是一个d维度的向量，梯度向量为$\nabla f(x) = [\frac{\partial f(x)}{\partial x_1}, \frac{\partial f(x)}{\partial x_2}, ..., \frac{\partial f(x)}{\partial x_d}]^T$。梯度向量中的每个偏导数其实表示f在x有关输入$x_i$的变化率

我们可以定义方向导数
$$D_uf(x) = \lim_{h \rightarrow 0} \frac{f(x+hu) - f(x)}{h} = \nabla f(x)u$$
方向导数$D_uf(x)$给出了f在x上沿着所有可能方向的变化率，我们希望最小化方向导数，即有$D_uf(x) = \parallel \nabla f(x) \parallel \parallel u \parallel cos(\theta)$，当$\theta = \pi$，方向导数最小，即沿着负梯度方向进行更新
$$x := x - \eta \nabla f(x)$$

```python
# 以2维为例
def train_2d(trainer, **kwargs):
    """
    function: 训练，并保存结果
    params trainer: 传入的是更新梯度的方式"""
    # 初始化
    x1, x2, s1, s2 = -5, -2, 0, 0
    # 存储结果
    result = [(x1, x2)]
    # 更新20次
    for i in range(20):
        # 使用trainer来进行更新
        x1, x2, s1, s2 = trainer(x1, x2, s1, s2, **kwargs)
        result.append((x1, x2))
    print(f"Epoch {i+1}, x1: {x1}, x2: {x2}")
    return result

def show_trace_2d(f, result):
    """
    function: 画轨迹图
    params result: 训练过程中保留的结果
    """
    plt.plot(*zip(*result), "-o", color="orange")
    x1, x2 = np.meshgrid(np.arange(-5.5, 1., 0.1), np.arange(-3., 1., 0.1))
    plt.contour(x1, x2, f(x1, x2))
    plt.xlabel("x1")
    plt.ylabel("x2")
```

以$f(x) = x_1^2 + 2x_2^2$

```python
def f_2d(x1, x2):
    """目标函数"""
    return x1**2 + 2*x2**2

def gd_2d(x1, x2, s1, s2, eta):
    x1 -= eta * 2 * x1
    x2 -= eta * 4 * x2
    return x1, x2, s1, s2
```

```python
result = train_2d(gd_2d, eta=0.1)
show_trace_2d(f_2d, result)
```

### 随机梯度下降

在深度学习中，目标函数通常用训练数据中有关各样本损失函数的平均，即$f(x) = \frac{1}{n} \sum_{i=1}^n f_i(x)$。目标函数在x处的梯度计算为$\nabla f(x) = \frac{1}{n} \sum_{i=1}^n \nabla f_i (x)$

当使用梯度下降时，每次更新的计算开销为`O(n)`,它随着n线性增长，当样本很大的时候，梯度下降每次迭代的计算开销很大。随机梯度下降则是随机均匀采样一个索引的梯度来更新，因此计算开销降为`O(1)`。
- 随机梯度$\nabla f_i(x)$时对梯度$\nabla f(x)$的无偏估计，平均来说，随机梯度是对梯度的一个良好估计

```python
def gd_2d(x1, x2, s1, s2, eta):
    x1 -= eta * (2 * x1 + np.random.normal(0, 1))
    x2 -= eta * (4 * x2 + np.random.normal(0, 1))
    return x1, x2, s1, s2

result = train_2d(gd_2d, eta=0.1)
show_trace_2d(f_2d, result)
```

### 小批量梯度下降

每轮迭代中随机均匀采样多个样本组成一个小批量，具体来说：
$$\nabla_{\beta} f(x) = \frac{1}{|\beta|} \sum_{i \in \beta_i} \nabla f_i (x_{t-1})$$
更新有$x_t := x_{t-1} - \eta_t g_t$
因为基于随机采样得到的梯度方差在迭代过程中无法减少，可以使用学习率衰减来使得学习率和梯度乘积的方差减少。而批量梯度下降由于一直使用目标函数真实的梯度，所以不需要用学习率衰减

- 批量小，并行效果不好，计算相同数量的样本需要更长的时间
- 批量大，信息冗余，获得更好的解要更多的样本

#### 数据读取

```python
import time
import torch
import torch.nn as nn
import torch.utils.data as Data

def load_data_airfoil():
    # 读取数据
    data = np.genfromtxt("../data/airfoil_self_noise.dat", delimiter="\t")
    # 标准化
    data = (data - data.mean()) / data.std()
    return torch.tensor(data[:1500, :-1], dtype=torch.float32), torch.tensor(data[:1500, -1], dtype=torch.float32)
```

```python
# feature and label
feature, label = load_data_airfoil()
feature.shape
label.shape
```

#### 从零实现

```python
def sgd(params, states, hyperparams):
    for param in params:
        param.data -= hyperparams["lr"] * param.grad.data
```

```python
def linreg(x, w, b):
    return torch.mm(x, w) + b


def squared_loss(y_pred, y):
    """
    calculate mean square loss which divide batch_size,
    and don't divide batch_size when update gradient by mini-batch GD.
    """
    return ((y_pred - y)**2).sum() /(len(y))
```

```python
def train_optim(net, loss, optimizer_fn, states, hyperparams, feature, label, batch_size, epoch_num):
    """
    function: train by various optimizer
    params optimizer_fn: optimzier
    params states: state input
    params hyperparams: such as lr
    """
    w = nn.Parameter(torch.normal(0, 0.01, size=(feature.shape[1], 1)), requires_grad=True)
    b = nn.Parameter(torch.zeros(1), requires_grad=True)
    
    # 生成data_iter(有3000个样本)
    dataset = Data.TensorDataset(feature, label)
    data_iter = Data.DataLoader(dataset, shuffle=True)

    # 测试误差
    def eval_loss():
        return loss(net(feature, w, b), label).item()
    loss_list = [eval_loss()]
    
    for i in range(epoch_num):
        start = time.time()
        for batch_i, (x, y) in enumerate(data_iter):
            l = loss(net(x, w, b), y)
            
            # grad clear
            if w.grad is not None:
                w.grad.data.zero_()
                b.grad.data.zero_()
                
            # grad backward
            l.backward()
            # update grad
            optimizer_fn([w, b], states, hyperparams)
            
            # save loss every 100 sample
            if (batch_i + 1) * batch_size % 100 == 0:
                loss_list.append(eval_loss())
    print(len(loss_list))         
    # draw（显示最后一个loss）
    print(f"loss: {loss_list[-1]}, {time.time() - start} sec per epoch")
    plt.plot(np.linspace(0, epoch_num, len(loss_list)), loss_list)
    plt.xlabel("epoch")
    plt.ylabel("loss")
```

```python
def train_sgd(lr, batch_size, epoch_num):
    train_optim(linreg, squared_loss, sgd, None, {"lr": lr}, feature, label, batch_size, epoch_num)
```

```python
# 批量梯度下降（学习率稍微放大一点）
train_sgd(0.01, 1500, 6)
```

```python
# 随机梯度下降
train_sgd(0.01, 1, 2)
```

```python
# 小批量梯度下降
train_sgd(0.01, 10, 2)
```

#### 简洁实现

```python
# optimizer_fn传入torch.optim.SGD等
def train_optim_pytorch(optimizer_fn, optimizer_hyperparams, feature, label, batch_size, epoch_num):
    # define network
    net = nn.Sequential(
        nn.Linear(feature.shape[-1], 1)
    )
    # define mse loss
    loss = nn.MSELoss()
    # define optimizer
    optimizer = optimizer_fn(net.parameters(), **optimizer_hyperparams)
    
    # 生成data_iter(有3000个样本)
    dataset = Data.TensorDataset(feature, label)
    data_iter = Data.DataLoader(dataset, shuffle=True)
    
    # 测试误差
    def eval_loss():
        return loss(net(feature).view(-1), label).item()
    
    loss_list = [eval_loss()]
    
    for _ in range(epoch_num):
        start = time.time()
        for batch_i, (x, y) in enumerate(data_iter):
            # calculate loss
            l = loss(net(x).view(-1), y)
            
            # grad clear
            optimizer.zero_grad()
            # backward()
            l.backward()
            # update grad
            optimizer.step()
            
            # save loss in every 100 example
            if (batch_i + 1) * batch_size % 100 == 0:
                loss_list.append(eval_loss())
                
    # draw
    print(f"loss: {loss_list[-1]}, {start - time.time()} sec per epoch")
    plt.plot(np.linspace(0, epoch_num, len(loss_list)), loss_list)
    plt.xlabel("epoch")
    plt.ylabel("loss")
```

```python
train_optim_pytorch(torch.optim.SGD, {"lr": 0.01}, feature, label, batch_size=10, epoch_num=2)
```

### 动量法

#### 梯度下降法的问题

当自变量迭代方向仅仅取决于自变量当前位置的时候，会带来一些问题

以$f(x) = 0.1x_1^2 + 2x_2^2$为例

```python
# 定义目标函数
def f_2d(x1, x2):
    return 0.1*x1**2 + 2*x2**2

def gd_2d(x1, x2, s1, s2, eta):
    x1 -= eta * 0.2 * x1
    x2 -= eta * 4 * x2
    return x1, x2, s1, s2

show_trace_2d(f_2d, train_2d(gd_2d, eta=0.4))
```

目标函数在x2方向上比x1方向上的斜率绝对值要更大。因此，给定固定的学习率时，梯度下降迭代会使得自变量在x2方向上移动幅度更大
- 此时，可能需要一个更小的学习率来避免x2方向上越过目标函数的最优解
- 这样往往会造成自变量在x1方向上的速度变慢

```python
# 学习过大时，在x2方向上不收敛
show_trace_2d(f_2d, train_2d(gd_2d, eta=0.6))
```

#### 动量法原理

为了解决上面的问题，动量法引入一个速度变量：
$$\Delta \theta_t = \rho \Delta \theta_{t-1} + g_t$$
$$\theta_t = \theta_{t-1} - \alpha \Delta \theta_t$$

```python
def momentum(x1, x2, v1, v2, gamma, eta):
    v1 = gamma * v1 + 0.2 * x1
    v2 = gamma * v2 + 4 * x2
    return x1 - eta * v1, x2 - eta * v2, v1, v2


show_trace_2d(f_2d, train_2d(momentum, gamma=0.5, eta=0.4))
```

由图上可以看到，动量法在x2轴方向上的移动更加平滑，并且在水平方向上更快逼近最优解，当学习率太大时，也不至于发散

```python
show_trace_2d(f_2d, train_2d(momentum, gamma=0.5, eta=0.6))
```

实际上，动量法就是在每个时间步的自变量更新量近似于将最近$1/(1-\rho)$个时间步的普通更新量（即学习率乘以梯度），做指数加权移动平均后再除以$1-\rho$
- 动量法，自变量在各个方向上的移动幅度不但取决于当前梯度，还取决于过去的各个梯度在各个方向上是否一致

#### 从零实现

```python
def init_momentum_state():
    # 初始化速度变量(非学习参数),同w和b同维度
    v_w = torch.zeros(feature.shape[1], 1)
    v_b = torch.zeros(1)
    return (v_w, v_b)

def sgd_momentum(params, states, hyperparams):
    # params是可学习参数，states是速度变量
    for p, v in zip(params, states):
        v.data = hyperparams["momentum"] * v.data + p.grad.data
        p.data -= hyperparams["lr"] * v.data
        
        
train_optim(linreg, squared_loss, sgd_momentum, init_momentum_state(), {"lr": 0.001, "momentum": 0.9}, 
            feature, label, batch_size=10, epoch_num=2)
```

#### 简洁实现

```python
train_optim_pytorch(torch.optim.SGD, {"lr": 0.001, "momentum": 0.9}, feature, label, batch_size=10, epoch_num=2)
```

### Adagrad

`Adagrad`是通过改变学习率来克服因为维度梯度值相差较大，所以引起的问题
- 为了使得自变量在梯度值较大的维度上不发散，需要选择较小的学习率
- 但是这样会导致自变量在梯度值较小的维度上迭代过慢

`Adagrad`就是根据自变量在每个维度的梯度值大小来调整各个维度上的学习率，从而避免统一的学习率难以适应所有维度的问题

$$G_t = G_{t-1} + g_t * g_t$$
$$\theta_t = \theta_{t-1} - \frac{\alpha}{\sqrt{G_t + \epsilon}} * g_t$$

这里的开方，除法和乘法运算都是按元素运算，目标函数自变量中每个元素都分别有自己的学习率
- 但是由于$G_t$一直在累加梯度，所以自变量中每个元素的学习率在迭代过程中一直在降低（或者不变）
- 因此，如果学习率在迭代早期降得较快，但是当前解依然不佳，那么由于`Adagrad`在迭代后期由于学习率过小，可能比较难找到一个有用的解

```python
def adagrad_2d(x1, x2, s1, s2, eta):
    g1, g2, eps = 0.2 * x1, 4 * x2, 1e-6
    s1 += g1 ** 2
    s2 += g2 ** 2
    x1 -= eta * np.sqrt(s1 + eps) * g1
    x2 -= eta * np.sqrt(s2 + eps) * g2
    return x1, x2, s1, s2

show_trace_2d(f_2d, train_2d(adagrad_2d, eta=0.03))
```

#### 从零实现

```python
def init_adagrad_state():
    sw = torch.ones(feature.shape[1], 1)
    sb = torch.zeros(1)
    return (sw, sb)

def adagrad(params, state, hyperparams):
    eps = 1e-6
    for p, s in zip(params, state):
        s.data += (p.grad.data ** 2)
        p.data -= hyperparams["lr"] * p.grad.data / torch.sqrt(s + eps)
        
train_optim(linreg, squared_loss, adagrad, init_adagrad_state(), {"lr": 0.01}, 
            feature, label, batch_size=10, epoch_num=2)
```

#### 简洁实现

```python
train_optim_pytorch(torch.optim.Adagrad, {"lr": 0.01}, feature, label, batch_size=10, epoch_num=2)
```

### PMSProp

`PMSProp`是`Adagrad`的改进版
- 因为$G_t$一直在累加梯度，所以学习率一直在下降或者不变
- `PMSProp`借鉴了移动平均的思想

$$G_t = \beta G_{t-1} + (1 - \beta)g_t*g_t$$
$$\theta_t = \theta_{t-1} - \frac{\alpha}{\sqrt{G_t + \epsilon}} * g_t$$

```python
def rmsprop(x1, x2, s1, s2, eta, beta):
    g1, g2, eps = 0.2*x1, 4*x2, 1e-6
    s1 = beta * s1 + (1 - beta) * (g1 ** 2)
    s2 = beta * s2 + (1 - beta) * (g2 ** 2)
    x1 -= eta * g1 / np.sqrt(s1 + eps)
    x2 -= eta * g2 / np.sqrt(s2 + eps)
    return x1, x2, s1, s2

show_trace_2d(f_2d, train_2d(rmsprop, eta=0.4, beta=0.9))
```

```python
def init_rmsprop_state():
    s_w = torch.ones(feature.shape[1], 1)
    s_b = torch.zeros(1)
    return s_w, s_b

def rmsprop(params, state, hyperparams):
    eps = 1e-6
    for p, s in zip(params, state):
        s.data = hyperparams["beta"] * s.data + (1 - hyperparams["beta"])*(p.grad.data ** 2)
        p.data -= hyperparams["lr"] * p.grad.data / torch.sqrt(s + eps)
             
train_optim(linreg, squared_loss, rmsprop, init_rmsprop_state(), {"lr": 0.01, "beta": 0.9}, 
            feature, label, batch_size=10, epoch_num=2)
```

#### 简洁实现

```python
train_optim_pytorch(torch.optim.RMSprop, {"lr": 0.01, "alpha": 0.9}, feature, label, batch_size=10, epoch_num=2)
```

### AdaDelta

`AdaDelta`是另外一种改进`Adagrad`的方式，但是它没有学习率这个参数，它除了像`RMSprop`引入移动平均，还使用过了一个额外的状态变$\Delta x_t$，用它来代替学习率

$$G_t = \beta G_{t-1} + (1 - \beta) g_t * g_t$$
$$g'_t = \sqrt{\frac{\Delta x_{t-1} + \epsilon}{G_t + \epsilon}} *g_t$$
$$\Delta x_t = \beta \Delta x_{t-1} + (1 - \beta) g'_t * g'_t$$
$$x_t = x_{t-1} - g'_t$$

`Adagrad`和`RMSProp`不同的是，使用了$\sqrt{\Delta_{t-1}}$来代替学习率$\eta$

```python
def init_adadelta_state():
    G_w = torch.ones(feature.shape[1], 1)
    G_b = torch.zeros(1)
    delta_w = torch.ones(feature.shape[1], 1)
    delta_b = torch.ones(1)
    return (G_w, delta_w), (G_b, delta_b)

def adadelta(params, state, hyperparams):
    beta, eps = hyperparams["beta"], 1e-6
    for p, (G, delta) in zip(params, state):
        G.data = beta * G.data + (1 - beta) * (p.grad.data ** 2)
        g = torch.sqrt((delta + eps)/(G + eps)) * p.grad.data
        p.data -= g
        delta.data = beta * delta.data + (1 - beta) * g * g
        
train_optim(linreg, squared_loss, adadelta, init_adadelta_state(), {"beta": 0.1}, 
            feature, label, batch_size=10, epoch_num=2)
```

#### 简洁实现

```python
train_optim_pytorch(torch.optim.Adadelta, {"rho": 0.9}, feature, label, batch_size=10, epoch_num=2)
```

### Adam

`Adam`是`RMSProp`和`Momentum`的结合，其在`RMSProp`的基础上对小批量随机梯度因为做了指数加权移动平均

$$G_t = \beta_1 G_{t-1} + (1 - \beta_1) g_t * g_t$$
$$\Delta \theta_t = \beta_2 \Delta \theta_{t-1} + (1 - \beta_2) g_t$$
$$G'_t = \frac{G_t}{1 - \beta_1^t}$$
$$\Delta \theta'_t = \frac{\Delta \theta_t}{1 - \beta_2^t}$$
$$\theta_{t} = \theta_{t-1} - \frac{\alpha}{\sqrt{G'_t + \epsilon}} \Delta \theta'_t$$

#### 从零实现

```python
def init_adam_state():
    G_w = torch.ones(feature.shape[1], 1)
    G_b = torch.zeros(1)
    delta_w = torch.ones(feature.shape[1], 1)
    delta_b = torch.ones(1)
    return (G_w, delta_w), (G_b, delta_b)

def adam(params, state, hyperparams):
    beta1, beta2, eps, t = 0.9, 0.999, 1e-6, 1
    for p, (G, delta) in zip(params, state):
        G.data = beta1 * G.data + (1 - beta1) * (p.grad.data ** 2)
        delta.data = beta2 * delta.data + (1 - beta2) * p.grad.data
        G_bias_corr = G / (1 - beta1 ** t)
        delta_bias_corr = delta / (1 - beta2 ** t)
        p.data -= hyperparams["lr"] * delta_bias_corr / torch.sqrt(G_bias_corr + eps)
        t += 1
        
train_optim(linreg, squared_loss, adam, init_adam_state(), {"lr": 0.001}, 
            feature, label, batch_size=1500, epoch_num=2)
```

#### 简洁实现

```python
train_optim_pytorch(torch.optim.Adam, {"lr": 0.01}, feature, label, batch_size=10, epoch_num=2)
```
