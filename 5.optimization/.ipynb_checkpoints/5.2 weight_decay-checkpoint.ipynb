## 权重衰减

在大多数框架下使用L2正则化来实现，但是L2正则化并不严格等同于权重衰减
- 在一些较为复杂的优化器中（像adam等），L2正则化不是权重衰减

L2正则化（在线性模型中又叫使用了L2的模型为岭回归模型）
- $L = \frac{1}{2N} \sum_{i=1}^{N} \parallel \hat y - y \parallel^2 + \frac{\lambda}{2N}w^Tw$
- 求导为$g_t + \lambda w$，$g_t$为未加正则的损失函数的梯度

权重衰减的原理
- $\theta_{t} = (1 - \beta)\theta_{t-1} - \alpha g_t$
- 实现的时候通常采用L2正则
    - $\theta_{t} = \theta_{t-1} - \alpha (g_t + \lambda \theta_{t-1})$
    - 在原理中正则部分$\theta_{t-1}$是除了batch_size的，但是在Pytorch中并没有除
    - pytorch默认对权重和偏置项进行权重衰减
    - 通过会使得学到的权重系数较为接近0

## 高维线性回归实验

设特征有200个，训练样本20个，测试数据有100个
- $y = 0.05 + \sum_{i=1}^p 0.01 x_i + \epsilon$

```python
import sys
import torch
import torch.nn as nn
import torch.utils.data as Data
sys.path.append("../d2l_func/")
from model_train import train_experiment
```

```python
# 生成数据
fea_num, train_num, test_num = 200, 20, 100
true_w = torch.full((fea_num, 1), 0.01)
true_b = torch.tensor([0.05])
x = torch.normal(0, 1, size=(train_num+test_num, fea_num))
error = torch.normal(0, 0.01, size=(len(x), 1))
y = torch.mm(x, true_w) + true_b + error
```

```python
class LinearModel(nn.Module):
    def __init__(self, fea_num):
        super(LinearModel, self).__init__()
        self.layer = nn.Sequential(
            nn.Linear(fea_num, 1)
        )
        
    def forward(self, x):
        return self.layer(x)
```

### 未添加权重衰减

```python
# define model
model = LinearModel(fea_num)
loss = nn.MSELoss()
train_dataset = Data.TensorDataset(x[:20], y[:20])
test_datatest = Data.TensorDataset(x[20:], y[20:])
test_iter = Data.DataLoader(test_datatest, batch_size=len(test_datatest), shuffle=True)

params = {
    "model": model,
    "loss": loss,
    "epoch_num": 100,
    "data_num": 20,
    "batch_size": 20,
    "test_iter": test_iter,
    "lr": 0.01,
    "weight_decay": 0,
    "draw": True,
}

# optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=params["lr"], weight_decay=params["weight_decay"])
train_iter = Data.DataLoader(train_dataset, batch_size=params["batch_size"], shuffle=True)
params["optimizer"] = optimizer
params["train_iter"] = train_iter

# training
train_experiment(**params)
```

### 添加权重衰减

```python
# define model
model = LinearModel(fea_num)
loss = nn.MSELoss()
train_dataset = Data.TensorDataset(x[:20], y[:20])
test_datatest = Data.TensorDataset(x[20:], y[20:])
test_iter = Data.DataLoader(test_datatest, batch_size=len(test_datatest), shuffle=True)

params = {
    "model": model,
    "loss": loss,
    "epoch_num": 100,
    "data_num": 20,
    "batch_size": 20,
    "test_iter": test_iter,
    "lr": 0.01,
    "weight_decay": 2,
    "draw": True,
}

# optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=params["lr"], weight_decay=params["weight_decay"])
train_iter = Data.DataLoader(train_dataset, batch_size=params["batch_size"], shuffle=True)
params["optimizer"] = optimizer
params["train_iter"] = train_iter

# training
train_experiment(**params)
```

可以看到测试误差也降下来了

### 个人实现

不调用`nn.Module`和优化器

```python
from optim import square_loss
```

```python
w = torch.normal(0, 0.01, size=(fea_num, 1), requires_grad=True)
b = torch.zeros([1], requires_grad=True)

def linreg(x):
    y_pred = torch.mm(x, w) + b
    return y_pred

net = linreg
# 这里实现的函数除了2，在pytorch中不除2
loss = square_loss
train_dataset = Data.TensorDataset(x[:20], y[:20])
test_datatest = Data.TensorDataset(x[20:], y[20:])
test_iter = Data.DataLoader(test_datatest, batch_size=len(test_datatest), shuffle=True)

params = {
    "model": net,
    "loss": loss,
    "epoch_num": 100,
    "data_num": 20,
    "batch_size": 20,
    "test_iter": test_iter,
    "params": [w, b], 
    "lr": 0.01,
    "weight_decay": 2,
    "draw": True,
}

train_iter = Data.DataLoader(train_dataset, batch_size=params["batch_size"], shuffle=True)
params["train_iter"] = train_iter

train_experiment(**params)
```

### 只对权重进行衰减

实际上L2正则化中，只对权重进行正则化，并没有对偏置项进行正则化
- 这里实验只对权重进行衰减

```python
net = LinearModel(fea_num)
loss = nn.MSELoss()
train_dataset = Data.TensorDataset(x[:20], y[:20])
test_datatest = Data.TensorDataset(x[20:], y[20:])
test_iter = Data.DataLoader(test_datatest, batch_size=len(test_datatest), shuffle=True)

params = {
    "model": net,
    "loss": loss,
    "epoch_num": 100,
    "data_num": 20,
    "batch_size": 20,
    "test_iter": test_iter,
    "params": [w, b], 
    "lr": 0.01,
    "weight_decay": 2,
    "draw": True,
}

train_iter = Data.DataLoader(train_dataset, batch_size=params["batch_size"], shuffle=True)
# 如果不写{"params": net.layer[0].bias}, 那在梯度更新时优化器将不会更新bias
optimizer = torch.optim.SGD([
    {"params": net.layer[0].weight, "weight_decay":2},
    {"params": net.layer[0].bias}
], lr=0.01)
params["optimizer"] = optimizer
params["train_iter"] = train_iter

# training
train_experiment(**params)
```

同样的参数下，只对权重衰减的效果比同时对权重和偏置项衰减的效果差
