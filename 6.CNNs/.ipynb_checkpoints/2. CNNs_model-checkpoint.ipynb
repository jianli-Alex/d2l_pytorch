## 全连接和卷积

- 全连接层参数多，运算量大，也容易过拟合，很难捕捉局部不变性
- 卷积层参数大大减少，共享参数，可以捕捉局部不变性。
    - 卷积层保留输入形状，使得图像的像素在高和宽两个方向上的相关性可以被识别
    - 卷积层通过滑动窗口将同一卷积核与不同的位置输入重复计算，从而避免参数尺寸过大

## LeNet5

总共有7层（3个卷积，2个池化，2个全连接）
- conv1: 1\*28\*28灰度图像（最初论文中其实是1\*32\*32的图像），经过6个5\*5的卷积核(padding为2)，变成了6\*28\*28的输出
    - 可学习参数有6*(5*5+1)=156个，连接数156\*28\*28=122304个
- pool1: 6\*28\*28的卷积层，经过2*2的最大池化，得到6*14*14
    - 可学习参数有6\*(1+1)=12个，连接数6\*(4+1)\*14\*14=5880个
- conv2: 6\*14\*14的最大汇聚层，经过96个5\*5的卷积核，变成16\*10\*10的卷积层
    - 论文中实际上使用连接表，用了60个5\*5的卷积核，先3个连续子集，然后4个连续子集，然后4个不连续子集，最后是所有的子集
    - 可学习参数96\*(5\*5)+16=2416个，连接数为16\*((6\*5\*5+1)\*10\*10)=241600个
- pool2: 16\*10\*10的卷积层经过2\*2的最大池化，变成16\*5\*5的输出
    - 可学习参数为16\*(1+1)=32个，连接数为16\*(4+1)\*5\*5=2000个
- conv3: 16\*5\*5的最大汇聚层，经过1920个5\*5的卷积核，变成120\*1\*1的输出
    - 可学习参数为1920\*5\*5+120=48120，连接数为120\*((16\*5\*5+1)\*1\*1)=48120个
- fc1: 120个神经元经过全连接变成84个神经元
    - 可学习参数和连接数都是120\*84+84=10164个
- fc2: 84个神经元经过全连接变成10个神经元
    - 可学习参数和连接数都是84\*10+10=850个
    
以前，LeNet5（论文中）也会在池化层后加上激活函数，但是现在一般不加

````python
import sys
import torch
import torch.nn as nn
import torch.nn.init as init
import torch.utils.data as Data
from collections import OrderedDict
sys.path.append("../d2l_func/")
from model_train import train_pytorch, train_epoch
from data_prepare import download_data_fashion_mnist, load_data_fashion_mnist
````

````python
class LeNet5(nn.Module):
    def __init__(self):
        super(LeNet5, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, padding=2),
            nn.ReLU(),
#             nn.Tanh(),
            nn.MaxPool2d(kernel_size=2),
            nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5),
            nn.ReLU(),
#             nn.Tanh(),
            nn.MaxPool2d(kernel_size=2),
            nn.Conv2d(in_channels=16, out_channels=120, kernel_size=5),
            nn.ReLU(),
#             nn.Tanh(),
        )
        
        self.fc = nn.Sequential(
            nn.Linear(120, 84),
            nn.ReLU(),
#             nn.Tanh(),
            nn.Linear(84, 10)
        )
        
    def forward(self, x):
        # 出来的是(batch, 120, 1, 1)，要将它变成向量
        output = self.conv(x)
        output = output.view(output.shape[0], -1)
        return self.fc(output)
    
    def score(self, x, y):
        y_pred = self.forward(x)
        acc = (y_pred.argmax(dim=1) == y).sum().item() / len(y)
        return acc
````

### optimizer实验

在常规写法中，`optimizer.zero_grad()`是放在`loss.backward()`前面的。而在梯度累加中给出的写法是，`optimizer.zero_grad()`放在`loss.backward()`后面
- 实验训练的第一次迭代中`optimizer.zero_grad`放在前面和放在后面有没有影响

````python
model = LeNet5()
loss = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
````

没有backward前的梯度

````python
train_iter, test_iter = load_data_fashion_mnist(batch_size=512)
x, y = iter(train_iter).next()
y_pred = model(x)
l = loss(y_pred, y)
````

````python
# count = 1
for param in optimizer.param_groups[0]["params"]:
#     print(count)
    print(param.shape)
    print(param.grad)
#     print("\n")
#     count += 1
    break
````

此时，可以看出在backward前，是没有梯度的

backward，并更新梯度

````python
l.backward()
optimizer.step()
````

````python
# 再看一下
for param in optimizer.param_groups[0]["params"]:
    print(param.shape)
    print(param.grad[0])
    break
````

变成有梯度了

````python
# 梯度清零
optimizer.zero_grad()
for param in optimizer.param_groups[0]["params"]:
    print(param.shape)
    print(param.grad[0])
    break
````

## AlexNet

AlexNet是第一现代的深层网络
- 总共包含了8层变换（5个卷积+3个fc）
    - 卷积-池化-卷积-池化-卷积-卷积-卷积-池化-fc-fc-fc
    - 相比了LeNet一直使用5x5的卷积核，AlexNet先是使用了11x11的卷积核，再使用5x5的卷积核，最后接连使用3x3的卷积核
- AlexNet将sigmoid改成了relu函数
    - relu没有饱和区域，单边抑制，计算速度快
- AlexNet使用了丢弃法来控制全连接层的复杂度
- Alex使用了大量的图像增广来进一步扩大数据集来缓解过拟合

AlexNet的结构

![AlexNet](./img/AlexNet.png)

````python
from collections import OrderedDict
````

````python
class AlexNet(nn.Module):
    def __init__(self):
        super(AlexNet, self).__init__()
        self.conv = nn.Sequential(OrderedDict({
            "conv1": nn.Conv2d(in_channels=1, out_channels=96, kernel_size=11, padding=3, stride=4),
            "relu1": nn.ReLU(),
            "pool1": nn.MaxPool2d(kernel_size=3, stride=2),
            "conv2": nn.Conv2d(in_channels=96, out_channels=256, kernel_size=5, padding=2, stride=1),
            "relu2": nn.ReLU(),
            "pool2": nn.MaxPool2d(kernel_size=3, stride=2),
            "conv3": nn.Conv2d(in_channels=256, out_channels=384, kernel_size=3, padding=1, stride=1),
            "relu3": nn.ReLU(),
            "conv4": nn.Conv2d(in_channels=384, out_channels=384, kernel_size=3, padding=1, stride=1),
            "relu4": nn.ReLU(),
            "conv5": nn.Conv2d(in_channels=384, out_channels=256, kernel_size=3, padding=1, stride=1),
            "relu5": nn.ReLU(),
            "pool3": nn.MaxPool2d(kernel_size=3, stride=2),
        }))
        
        self.fc = nn.Sequential(OrderedDict({
            "fc1": nn.Linear(256*6*6, 4096),
            "relu1": nn.ReLU(),
            "dropout1": nn.Dropout(0.5),
            "fc2": nn.Linear(4096, 4096),
            "relu2": nn.ReLU(),
            "dropout2": nn.Dropout(0.5),
            "fc3": nn.Linear(4096, 10)
        }))
        
    def forward(self, x):
        feature = self.conv(x)
        feature = feature.view(x.shape[0], -1)
        return self.fc(feature)
    
    def score(self, x, y):
        y_pred = self.forward(x)
        acc = (y_pred.argmax(dim=1) == y).sum().item() / len(y)
        return acc
````

### 梯度累加实验

梯度累加的通常写法：
```
for i, (image, label) in enumerate(train_loader):
    image, label = image.cuda(), label.cuda()
    y_pred = model(image)
    l = loss(y_pred, y) / accum_step
    l.backward()
    
    if i%accum_step == 0:
        optimizer.step()
        optimizer.zero_grad()
```

````python
torch.cuda.empty_cache()
````

````python
model = LeNet5()
model = model.cuda()
loss = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)
````

````python
mnist_train, mnist_test = download_data_fashion_mnist()

train_data, train_label = mnist_train.data[:100], mnist_train.targets[:100]
train_data = train_data.float()
dataset = Data.TensorDataset(train_data.view(100, -1, 28, 28), train_label)
train_iter = Data.DataLoader(dataset, batch_size=2)
````

````python
accum_step = 4
for epoch in range(10):
    for i, (x, y) in enumerate(train_iter):
        x = x.cuda()
        y = y.cuda()
        y_pred = model(x)
        l = loss(y_pred, y)/accum_step
        
        l.backward()
        if (i+1)%accum_step == 0:
            optimizer.step()
            optimizer.zero_grad()
        
        print(l)
    for param in optimizer.param_groups[0]["params"]:
        print(param.shape)
        print(param.grad)
        break
    print("\n")
````

如果在每一epoch结束后，不清空梯度，那么梯度将会累加到下一类，下一轮的前几次迭代就不应该除以accum_step了

## VGG

VGG是第一个真正意义上的深层网络，但是它的网络结构很简单就是不断堆积卷积和池化

````python
from collections import OrderedDict
from torchsummary import summary

def vgg_block(conv_num, in_channels, out_channels):
    layer = OrderedDict({})
    layer["conv1"] = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, stride=1)
    
    for i in range(1, conv_num):
        layer["conv" + str(i+1)] = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, stride=1)
        layer["relu" + str(i+1)] = nn.ReLU()
        
    layer["pool1"] = nn.MaxPool2d(kernel_size=2, stride=2)
    
    return nn.Sequential(layer)
````

````python
class VGG11(nn.Module):
    def __init__(self):
        super(VGG11, self).__init__()
        self.conv = nn.Sequential(
            vgg_block(1, 1, 64),
            vgg_block(1, 64, 128),
            vgg_block(2, 128, 256),
            vgg_block(2, 256, 512),
            vgg_block(2, 512, 512),
        )
        
        self.fc = nn.Sequential(
            nn.Linear(512*7*7, 4096),
            nn.Dropout(0.5),
            nn.Linear(4096, 4096),
            nn.Dropout(0.5),
            nn.Linear(4096, 10)
        )
        
    def forward(self, x):
        x = x.float()
        output = self.conv(x)
        output = output.view(output.shape[0], -1)
        return self.fc(output)
    
    def score(self, x, y):
        y_pred = model(x)
        acc = (y_pred.argmax(dim=1) == y).sum().item() / len(y)
        return acc
````

````python
class VGG16(nn.Module):
    def __init__(self):
        super(VGG16, self).__init__()
        self.conv = nn.Sequential(
            vgg_block(2, 1, 64),
            vgg_block(2, 64, 128),
            vgg_block(2, 128, 256),
            vgg_block(3, 256, 512),
            vgg_block(3, 512, 512),
        )
        
        self.fc = nn.Sequential(
            nn.Linear(512*7*7, 4096),
            nn.Dropout(0.5),
            nn.Linear(4096, 4096),
            nn.Dropout(0.5),
            nn.Linear(4096, 10)
        )
        
    def forward(self, x):
        output = self.conv(x)
        output = output.view(output.shape[0], -1)
        return self.fc(output)
    
    def score(self, x, y):
        y_pred = model(x)
        acc = (y_pred.argmax(dim=1) == y).sum().item() / len(y)
        return acc
````
